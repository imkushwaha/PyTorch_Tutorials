```python
# tensor operation

x = torch.tensor(3.)
w = torch.tensor(4., requires_grad=True)  # requires_grad = True means it will calculate gradient automatic for weight
b = torch.tensor(5., requires_grad=True)

x, w, b
```

```python
## Arithmetic operation

y = w*x + b
y
```

```python
## to calculate gradient

y.backward()
```

```python
print('dy/dx: ', x.grad)
print('dy/dw: ', w.grad)
print('dy/db: ', b.grad)
```
